```python
from collections import Counter
def load_grh(filepath):
    """Load a .grh PCE graph into NetworkX"""
    G = nx.Graph()
    with open(filepath, "r") as f:
        lines = f.readlines()

    n_nodes = len(lines)
    for u in range(n_nodes):
        G.add_node(u)

    for u, line in enumerate(lines):
        line = line.strip()
        neighbors = [int(x) for x in re.findall(r'\d+', line)]
        for v in neighbors:
            G.add_edge(u, v)
    G.remove_edges_from(nx.selfloop_edges(G))
    return G




# def compute_avg_core(graph_folder, filenames):
#     for filename in filenames:
#         path = os.path.join(graph_folder, filename)
#         print("Processing:", filename)

#         G = load_grh(path)
#         core_numbers = nx.core_number(G)
#         total_nodes = len(core_numbers)
#         core_counts = Counter(core_numbers.values())
        
#         print(core_numbers)
#         print(total_nodes)
#         print(core_counts)
```

```python
def compute_avg_core(graph_folder, filenames):
    results = []
    i=1
    for filename in filenames:
        path = os.path.join(graph_folder, filename)
        print(f"{i} processing: {filename}")
        
        try:
            G = load_grh(path)
            core_numbers = nx.core_number(G)
            core_values = list(core_numbers.values())
                
            total_nodes = G.number_of_nodes()
            avg = sum(core_numbers.values())/total_nodes
            std_core = np.std(core_values) 
            print(f"""Total Nodes: {total_nodes}, avg: {avg}, std: {std_core}""")
#             print(core_values)
            
            results.append({
                "Graph" : filename,
                "Node" : total_nodes,
                "avg_core" : avg,
                "std_core" : std_core
            })
        except Exception as e:
            results.append({
                "Graph" : filename,
                "Node" : "",
                "avg_core" : "",
                "std_core" : "",
                "Error" : str(e)
            })
        i = i+1
            
    return pd.DataFrame(results)
```
