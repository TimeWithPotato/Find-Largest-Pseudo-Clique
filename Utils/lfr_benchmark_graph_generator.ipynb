```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.LFR_benchmark_graph(n=250, tau1=3, tau2= 1.5, mu = 0.1, average_degree = 5, min_community=10)
```

```python
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(G, seed=57, scale=1.5)
nx.draw(G, pos, node_size=30, node_color='lightcoral', edge_color='gray', with_labels=False)
plt.title("LFR Graph")
plt.show()
```
