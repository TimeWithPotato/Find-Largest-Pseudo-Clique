```python
# Maximum number of files to process per batch

# BATCH_SIZE = 3  # Max files per batch

# ------------------- Graph Reader -------------------
def load_grh(filepath):
   
    G = nx.Graph()
    with open(filepath, "r") as f:
        lines = f.readlines()

    n_nodes = len(lines)
    for u in range(n_nodes):
        G.add_node(u)

    for u, line in enumerate(lines):
        line = line.strip()
        neighbors = [int(x) for x in re.findall(r'\d+', line)]
        for v in neighbors:
            G.add_edge(u, v)
    G.remove_edges_from(nx.selfloop_edges(G))
    
    return G
```

```python
# ------------------- Max Core Processor -------------------
def compute_mxc_std_deg(graphf_folder, filenames):
    
    results = []
    i=1
    for filename in filenames:
        path = os.path.join(graph_folder, filename)
        print(f"{i} Processing: {filename}")
        try:
            G = load_grh(path)

            # Compute core numbers
            core_numbers = nx.core_number(G)
            mxc_num = max(core_numbers.values())
            # Extract max-core subgraph
            mxc_nodes = [n for n, k in core_numbers.items() if k == mxc_num]
            mxc_subgraph = G.subgraph(mxc_nodes).copy()
            
            print(mxc_subgraph)
            # Compute stats
            mxc_nodes = mxc_subgraph.number_of_nodes()
            degrees = dict(mxc_subgraph.degree())
            mxc_std_degree = np.std(list(degrees.values())) if mxc_nodes > 0 else 0
            print(mxc_std_degree)
            
            results.append({
                "Graph": filename,
                "mxc_std_degree" : mxc_std_degree
            })
        except Exception as e:
            results.append({
                "Graph": filename,
                "std_degree":"",
                "Error": str(e)
            })
        i = i+1
    return pd.DataFrame(results)
```
