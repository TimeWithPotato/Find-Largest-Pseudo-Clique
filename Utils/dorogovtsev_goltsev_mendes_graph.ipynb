```python
import networkx as nx
import matplotlib.pyplot as plt
```

nodes = 3/2 * (3**(n-1) + 1 and edges = 3**n

```python
n = 3
G = nx.dorogovtsev_goltsev_mendes_graph(n)
isConnected = nx.is_connected(G)
print(f"{'Connected' if isConnected else 'Not Connected'}")
```

```python
if isConnected:
    plt.figure(figsize=(6, 6))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G,pos, with_labels=True, node_color="Green", edge_color="gray")
    plt.title("Dorogovtsev-Goltsev-Mendes Graph (n=3)")
    plt.show()
```
